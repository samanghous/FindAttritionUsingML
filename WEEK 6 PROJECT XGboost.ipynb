```python
from sklearn.model_selection import RandomizedSearchCV


sam_hyperparam={
     
   'booster':['gbtree','dart'],  
   'max_depth':[None,5,6,7,8,9,12 ],
   'subsample':[0.5,0.6,0.7,0.74,0.79,0.84],
    
   'colsample_bytree':[ 0.85,0.9,0.93,0.97,1],
     
    'n_estimators':[500, 510, 520, 530, 540, 550, 560, 570, 580, 590, 600, 610, 620,
       630, 640, 650, 660, 670, 680, 690, 700, 710, 720, 730, 740, 750,
       760, 770, 780, 790, 800, 810, 820, 830, 840, 850, 860, 870, 880,
       890, 900, 910, 920, 930, 940, 950, 960, 970, 980, 990], 
    'learning_rate':[0.03],
     
} 
     



def sam_hyperpm_func(itern):
    from sklearn.model_selection import RandomizedSearchCV
    sam_bestparam=RandomizedSearchCV(sammodel,param_distributions=sam_hyperparam,n_jobs=-1,n_iter=itern,cv=4)
    sam_bestparam.fit(dffinal.drop(['Id','Attrition'],axis=1),dffinal['Attrition'])
    gh1=sam_bestparam.best_params_ 
    gh2=sam_bestparam.best_score_
    return gh1,gh2
```
