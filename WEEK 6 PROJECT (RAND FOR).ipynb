```python
from sklearn.model_selection import RandomizedSearchCV

from sklearn.ensemble import RandomForestClassifier
sammodel=RandomForestClassifier(n_jobs=-1)
```

```python
sam_hyperparam={
    'n_estimators':[70,72,74,76,77,79,80,82,83,85,87,88,100,200,300,400,600,800,1000],
    'max_depth':[None,67,71,74,76,79,80,81,82,84,85,87,88,89,90,93,95],
    'min_samples_split':[2,3,4,5,6,7],
    'min_samples_leaf':[2,3,4,5],
    'bootstrap':[False,True],
    'max_features':['auto','sqrt','log2',4,5,6,7,8],
    'class_weight':["balanced","balanced_subsample"]
} 
     



def sam_hyperpm_func(itern):
    from sklearn.model_selection import RandomizedSearchCV
    sam_bestparam=RandomizedSearchCV(sammodel,param_distributions=sam_hyperparam,n_jobs=-1,n_iter=itern,cv=3)
    sam_bestparam.fit(dffinal.drop(['Id','Attrition'],axis=1),dffinal['Attrition'])
    gh1=sam_bestparam.best_params_ 
    gh2=sam_bestparam.best_score_
    return gh1,gh2
```
